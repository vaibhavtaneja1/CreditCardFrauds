# Resampling to get rid of imbalanced data

We can either do **undersampling** of data or **oversampling** of data. What happens in both of them is described below

<ul>
    <li><b>UnderSampling:</b> we take ramdom draws from non-fraud observations to match the number of fraud observations. But we randomly loose alot of information in this</li>
    <li><b>OverSampling:</b> Random oversampling duplicates examples from the minority class in the training dataset and can result in overfitting for some models</li>
</ul>

Apart from these two, we have another method, which is proved more effective as compared to other two mentioned.
<ul>
<b>(SMOTE) Synthetic Minority Oversampling Technique:</b> It overcomes the data imbalance by oversampling the minority class (fraud cases) using nearest neighbors of fraud cases to create new synthetic fraud cases instead of duplicating the monority samples </ul>

```python
x = df_new.drop(['Class'], axis = 1).values
y = df_new.loc[:,['Class']].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import RandomOverSampler
from imblearn.over_sampling import SMOTE
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.metrics import roc_curve,roc_auc_score, precision_recall_curve, average_precision_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
import warnings
warnings.filterwarnings('ignore')

model = LogisticRegression()
model.fit(x_train, y_train)

# Obtain model predictions
y_pred = model.predict(x_test)
probs = model.predict_proba(x_test)
```

```python
fpr, tpr, threshold = roc_curve(y_test, y_pred)

# Calculate AUC
aucScore = roc_auc_score(y_test, probs[:, 1])
 
precision, recall, thresholds = precision_recall_curve(y_test, y_pred)

average_precision = average_precision_score(y_test, y_pred)
```

```python
def build_roc_curve(fpr,tpr,aucScore):
    fig = plt.figure(figsize = (15,5))
    ax = fig.add_subplot(111)
    ax.plot(fpr, tpr, label='AUC = %0.3f'% aucScore)
    ax.plot([0,1],[0,1])
    ax.set_title('Receiver operating characteristic curve (ROC)')
    ax.legend()
    ax.set_ylabel('True Positive Rate')
    ax.set_xlabel('False Positive Rate')
    plt.show()

# Define a precision_recall_curve function
def plot_pr_curve(recall, precision, average_precision):
    plt.step(recall, precision, color='b', alpha=0.2, where='post')
    plt.fill_between(recall, precision, step='post', alpha=0.2, color='b')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.ylim([0.0, 1.05])
    plt.xlim([0.0, 1.0])
    plt.title('2-class Precision-Recall curve: AP={0:0.2f}'.format(average_precision))
    plt.show()
```

```python
print('Classification report:\n', classification_report(y_test, y_pred))
print('Confusion matrix:\n',confusion_matrix(y_true = y_test, y_pred = y_pred))

# Plot the roc curve 
build_roc_curve(fpr,tpr,aucScore)

# Plot recall precision curve
plot_pr_curve(recall, precision, average_precision)

prn, rcl = precision[1], recall[1]
print('Recall:', rcl)
print('Precision:', prn)
print('F-Score:', (2*prn*rcl)/(prn+rcl))
```

___________________________________________________________________________________________________________________
Terms:
<ol>
    <li><b>True Positives:</b> Correctly Classified Fraud Transactions </li>
    <li><b>False Positives:</b> Transactions which are not fraud but classified as fraud</li>
    <li><b>True Negative:</b> Correctly Classified Non-Fraud Transactions</li>
    <li><b>False Negative:</b> Transactions which are fraud but classified as not fraud</li>
    <li><b>Precision (1-Specificty):</b> TP/(TP + FP) i.e ratio of correct predicted positive to the total predicted positive </li>
    <li><b>Recall or Sensitivity:</b> TP/(TP + FN) i.e what percentage of fraud is correctly identified</li>
    <li><b>F1 Score</b> = 2(Recall. Precision) / (Recall + Precision) is the weighted average of Precision and Recall</li>
</ol>

**Precision** as the name suggests, tells us how precise (how sure) is our model in detecting fraud transactions
**Recall** is the amount of fraud cases our model is able to detect.

**Precision/Recall Tradeoff:** The more precise (selective) our model is, the less cases it will detect. 
Example: Assuming that our model has a precision of 95%, Let's say there are only 5 fraud cases in which the model is 95% precise or more that these are fraud cases. Then let's say there are 5 more cases that our model considers 90% to be a fraud case, if we lower the precision there are more cases that our model will be able to detect.
___________________________________________________________________________________________________________________

```python
sampling_dict = {'UnderSamp': RandomUnderSampler(random_state=0)
                ,'OverSamp' : RandomOverSampler(random_state=0)
                ,'SMOT': SMOTE(random_state=0)
                }
```

```python
for s in sampling_dict.keys():
    sample = sampling_dict[s]
    x_train_smpl, y_train_smpl = sample.fit_sample(x_train, y_train)
    model = LogisticRegression()
    model.fit(x_train_smpl, y_train_smpl)

    y_pred = model.predict(x_test)
    probs = model.predict_proba(x_test)

    aucScore = roc_auc_score(y_test, probs[:, 1])
 
    precision, recall, thresholds = precision_recall_curve(y_test, y_pred)
    
    prn, rcl = precision[1], recall[1]
    print('Recall for',s,'is:', rcl)
    print('Precision for',s,'is:',prn)
    print('F-Score for',s,'is:', (2*prn*rcl)/(prn+rcl))
    print('_'*50)
    x_train_smpl, y_train_smpl = 0,0
```

### It is important here that, not in all cases resampling can help us with achieving high precision and high recall
